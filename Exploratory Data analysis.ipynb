```python
# Count plot for categorical feature
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x = 'SaleCondn', data = df)
plt.title('Count of observations per Salescondition')
plt.xlabel('SaleCondn')
plt.ylabel('Count')
plt.show()
```

```python
# When the sale condition is normal than the highest houses count will sale
```

```python
plt.figure(figsize = (10,7))
sns.countplot(y='Neighborhood', data = df)
plt.title('Neighborhood vs Salecondition')
plt.xlabel('SaleCondn')
plt.ylabel('Neighborhood')
plt.show()
```

```python
#IN this graph we can see Neighborhood vs sale condition
#NAmes with in neighborhood is the highest in sale condition
```

```python
#Plot histogram for "Year Built","LivGriv Area" and PropPrice
```

```python
num_cols = ['GrLivArea','YearBuilt','OverallCond','PropPrice']
for i in range (0,len(num_cols),2):
    plt.figure(figsize=(10,4))
    plt.subplot(121)
    sns.distplot(df[num_cols[i]],kde=False)
    plt.subplot(122)
    sns.distplot(df[num_cols[i+1]],kde=False)
    plt.tight_layout()
    plt.show()
    
```

```python
#Here we can observe that the histogram plot is left skewed and right skewed
```

```python
##KDE = Kernel Density Estimation
```

```python
num_cols = ['GrLivArea','YearBuilt','OverallCond','PropPrice']
for i in range(0,len(num_cols),2):
    plt.figure(figsize = (10,6))
    plt.subplot(121)
    sns.distplot(df[num_cols[i]],hist = True, kde = True)
    plt.subplot(122)
    sns.distplot(df[num_cols[i+1]],hist = True, kde = True)
    plt.tight_layout()
    plt.show()
```

```python
#Barchart 
```

```python
plt.figure(figsize = (10,6),dpi = 200)
df['OverallQual'].value_counts().plot.bar(color = 'green', ls = '-',lw = '3')
plt.xlabel('OverallQual', color = 'red')
plt.ylabel('value_count')
plt.title('Bargraph on overallqual')
plt.legend()
plt.show()
```

```python
#Vertical Violin Plot
```

```python
sns.set(style="whitegrid")
f,ax = plt.subplots(figsize = (15,10))
#Show each distribution with both violins and points
sns.violinplot(x="SaleYr", y = "PropPrice", data = df, inner = 'box',palette="Set2",cut=2,linewidth=3,ax =ax)
sns.despine(left = True)
ax.set_title('SaleYear vs PropertyPrice')
ax.set_xlabel('SaleYr')
ax.set_ylabel('PropPrice')
```

```python
##Scatter PLOT
```

```python
sns.set(style = 'whitegrid')
f,ax = plt.subplots(figsize = (15,10))
sns.scatterplot(data = df,x = "SaleCondn", y = "PropPrice",legend = False,sizes = (20,2000),ax=ax)
ax.set_title('SaleCondition vs PropPrice')
ax.set_xlabel('SaleCondn')
ax.set_ylabel('PropPrice')
```

```python
#Bubble Plot
#A bubble plot is an extension of a scatter plot and replace the dots with bubbles.It is used to investigate between three variables 
```

```python
sns.set(style = 'whitegrid')
f,ax = plt.subplots(figsize = (12,8))
sns.scatterplot(data=df, x = 'SaleCondn', y = 'PropPrice', size='Neighborhood', hue='Neighborhood', alpha = 0.5,legend = False)
ax.set_title('salecondition vs Property Price for different Neighborhood')
ax.set_xlabel('SaleCondn')
ax.set_ylabel('PropPrice')
```
